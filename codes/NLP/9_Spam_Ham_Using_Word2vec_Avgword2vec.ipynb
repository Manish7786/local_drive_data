```python
import pandas as pd
messages = pd.read_csv('SpamCollectionSMS.txt',header=None,sep="\t", names=['label', 'message'])
```

```python
messages
```

```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
## Lets train Word2vec from scratch
model = gensim.models.Word2Vec(sentences=words, vector_size=100, window=5)
```

```python
## To get all the vocabulary
model.wv.index_to_key
```

```python
import numpy as np
def avg_word2vec(doc):
    vecs = [model.wv[word] for word in doc if word in model.wv]
    if len(vecs) > 0:
        return np.mean(vecs, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
from tqdm import tqdm
```

```python
## Apply for entire sentences
X = [avg_word2vec(doc) for doc in tqdm(words)]
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```
