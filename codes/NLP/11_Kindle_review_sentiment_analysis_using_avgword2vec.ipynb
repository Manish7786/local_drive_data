```python
## Load the Dataset
import pandas as pd
data = pd.read_csv("all_kindle_review .csv")
data.head()
```

```python
df = data[['reviewText', 'rating']]
df.head()
```

```python
## Positive review is 1 and negative review is 0
df['rating'] = df['rating'].apply(lambda x: 0 if x<3 else 1)
```

```python
import re
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
```

```python
## Removing special characters
df['reviewText'] = df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+', '', x))
## Remove the stopwords
df['reviewText'] = df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', str(x)))
## Remove html tags
df['reviewText'] = df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
## Remove any addtional spaces
df['reviewText'] = df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = df['reviewText'][i].split()
    review = [lemmatizer.lemmatize(word) for word in review]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
import gensim
```

```python
## Lets train Word2vec from scratch
model = gensim.models.Word2Vec(sentences=words, vector_size=100, window=5)
```

```python
## To get all the vocabulary
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
import numpy as np
def avg_word2vec(doc):
    vecs = [model.wv[word] for word in doc if word in model.wv]
    if len(vecs) > 0:
        return np.mean(vecs, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python
from tqdm import tqdm
```

```python
## Apply for entire sentences
X = [avg_word2vec(doc) for doc in tqdm(words)]
```

```python
def predict_review_sentiment(review_text):
    # Lowercase
    review_text = review_text.lower()
    
    # Remove special characters
    review_text = re.sub('[^a-z A-Z 0-9]+', '', review_text)
    
    # Remove stopwords
    review_text = " ".join([word for word in review_text.split() if word not in stopwords.words('english')])
    
    # Remove URL
    review_text = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', review_text)
    
    # Remove HTML
    review_text = BeautifulSoup(review_text, 'lxml').get_text()
    
    # Lemmatize
    lemmatized = [lemmatizer.lemmatize(word) for word in review_text.split()]
    
    # Convert to word2vec vector
    vec = avg_word2vec(lemmatized).reshape(1, -1)
    
    # Predict
    prediction = nb_model_avgword2vec.predict(vec)
    
    if prediction[0] == 1:
        return "Positive Review ✅"
    else:
        return "Negative Review ❌"
```

```python
test_review = "One of the best books I've read this year. Beautifully written."
print(predict_review_sentiment(test_review))
```
