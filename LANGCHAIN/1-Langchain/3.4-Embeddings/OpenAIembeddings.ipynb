```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "Hii My name is Manish Chauhan, I work as data scientist."

query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
text = "Hii My name is Manish Chauhan, I work as data scientist."

query_result = embeddings1024.embed_query(text)
len(query_result)
```

```python
## Text Loader.
from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
text = text_splitter.split_documents(docs)
text
```

```python
## Vector Embedding and vector StoreDB

from langchain.vectorstores import Chroma

db = Chroma.from_documents(text, embeddings1024)
db
```

```python
## Retrieve the results from the query vector storeDB.

query = "Bangkok traces its roots to a small trading post during"
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
