```python
## Feature Engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(np.int64)
df['Month'] = df['Month'].astype(np.int64)
df['Year'] = df['Year'].astype(np.int64)
```

```python
df.info()
```

```python
## Drop date of journey
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(np.int64)
df['Arrival_min'] = df['Arrival_min'].astype(np.int64)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
df['Departure_hour'] = df['Departure_hour'].astype(np.int64)
df['Departure_min'] = df['Departure_min'].astype(np.int64)
```
